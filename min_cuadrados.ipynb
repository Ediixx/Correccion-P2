La ecuación de la línea es:

$ y(x) = a_1 x + a_0 $

Al realizar el proceso de mínimos cuadrados queda el siguiente sistema de ecuaciones:


$( \sum_i (y_i - a_1 x_i - a_0) , \sum_i (y_i - a_1 x_i - a_0)x_i ) = 0$

## Conjunto de datos 1

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs1 = [
    -5.0000,
    -3.8889,
    -2.7778,
    -1.6667,
    -0.5556,
    0.5556,
    1.6667,
    2.7778,
    3.8889,
    5.0000,
]
ys1 = [
    57.2441,
    33.0303,
    16.4817,
    7.0299,
    0.5498,
    0.7117,
    3.4185,
    12.1767,
    24.9167,
    44.2495,
]
```

```python
plt.scatter(xs1, ys1)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Conjunto de datos 1")
plt.show()
```

Interpole el conjunto de datos 1 usando la función cuadrática. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Derivadas parciales para regresión cuadrática
def der_parcial_2(xs: list, ys: list) -> tuple[float, float, float, float]:
    der_2 = sum(xi**2 for xi in xs)  # Derivada parcial con respecto a a2
    der_1 = sum(xs)  # Derivada parcial con respecto a a1
    der_0 = len(xs)  # Derivada parcial con respecto a a0
    der_ind = sum(yi for yi in ys)  # Derivada parcial con respecto al término independiente
    return (der_2, der_1, der_0, der_ind)

def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float, float]:
    der_2 = sum(xi**3 for xi in xs)  # Derivada parcial con respecto a a2
    der_1 = sum(xi**2 for xi in xs)  # Derivada parcial con respecto a a1
    der_0 = sum(xs)  # Derivada parcial con respecto a a0
    der_ind = sum(xi * yi for xi, yi in zip(xs, ys))  # Derivada parcial con respecto al término independiente
    return (der_2, der_1, der_0, der_ind)

def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float, float]:
    der_2 = sum(xi**4 for xi in xs)  # Derivada parcial con respecto a a2
    der_1 = sum(xi**3 for xi in xs)  # Derivada parcial con respecto a a1
    der_0 = sum(xi**2 for xi in xs)  # Derivada parcial con respecto a a0
    der_ind = sum(xi**2 * yi for xi, yi in zip(xs, ys))  # Derivada parcial con respecto al término independiente
    return (der_2, der_1, der_0, der_ind)

def ajustar_min_cuadrados(xs: list, ys: list, gradiente: list) -> tuple[float, float, float]:
    A = []
    B = []
    for der in gradiente:
        der_2, der_1, der_0, der_ind = der(xs, ys)
        A.append([der_2, der_1, der_0])
        B.append(der_ind)
    coeffs = np.linalg.solve(A, B)
    return coeffs

# Ajustar una función cuadrática
a2, a1, a0 = ajustar_min_cuadrados(xs1, ys1, gradiente=[der_parcial_0, der_parcial_1, der_parcial_2])

# Mostrar los resultados
print(f"Coeficiente a2: {round(a2, 5)}")
print(f"Coeficiente a1: {round(a1, 5)}")
print(f"Coeficiente a0: {round(a0, 5)}")

# Calcular y(2.25)
x_val_1 = 2.25
y_val_1 = round(a2 * x_val_1**2 + a1 * x_val_1 + a0, 5)
print(f"y(2.25) = {y_val_1}")

# Calcular y(-2.25)
x_val_2 = -2.25
y_val_2 = round(a2 * x_val_2**2 + a1 * x_val_2 + a0, 5)
print(f"y(-2.25) = {y_val_2}")

# Graficar los datos y la curva de ajuste cuadrático
plt.scatter(xs1, ys1, label="Datos",color='red')
x_line = np.linspace(min(xs1), max(xs1), 100)
y_line = a2 * x_line**2 + a1 * x_line + a0
plt.plot(x_line, y_line, color='green', label="Ajuste cuadrático")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste por mínimos cuadrados")
plt.legend()
plt.show()
```

Interpole el conjunto de datos 2 usando la función exponencial. 

```python
# Transformar ys2 usando logaritmo natural
log_ys2 = np.log(ys2)

# Ajustar una línea recta a los datos transformados
coeffs = np.polyfit(xs2, log_ys2, 1)
b, log_a = coeffs
a = np.exp(log_a)

# Mostrar los resultados
print(f"Coeficiente a: {a}")
print(f"Coeficiente b: {b}")

# Calcular y(5) y y(1)
y_5 = a * np.exp(b * 5)
y_1 = a * np.exp(b * 1)

print(f"y(5) = {y_5}")
print(f"y(1) = {y_1}")

# Graficar los datos y la curva de ajuste exponencial
plt.scatter(xs2, ys2, color='skyblue', label="Datos")
x_line = np.linspace(min(xs2), max(xs2), 100)
y_line = a * np.exp(b * x_line)
plt.plot(x_line, y_line, color='brown', label="Ajuste exponencial")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Ajuste por mínimos cuadrados")
plt.legend()
plt.show()
```
